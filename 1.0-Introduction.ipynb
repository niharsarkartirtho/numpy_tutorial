# Creating arrays from python list/tuples
### array() function
### default values --> zeros function -->np.zeros(shape)
### default 1 values --> ones function ...tuples np.ones((2,3))
### full function---> specific value--->((shape),value)

```python
# array function ---> array()
arr2 = np.array(list1)
print(arr2)
```

```python
# default values
# np.zeros(shape)   // 1d-->(3) 2d--> (3,3)
zeros_array = np.zeros(2)
print(zeros_array)
```

```python
#ones funtion
ones_array = np.ones((2,3))
print(ones_array)
```

```python
#full function
filled_array = np.full(3,7)
filled_array2 = np.full((3,2),8)
print(filled_array)
print(filled_array2)
```

### How To Create Your Own ufunc
To create your own ufunc, you have to define a function, like you do with normal functions in Python, then you add it to your NumPy ufunc library with the frompyfunc() method.

The frompyfunc() method takes the following arguments:

function - the name of the function.  
 inputs - the number of input arguments (arrays).  
 outputs - the number of output arrays.

```python
def adding(x , y):
    return x+y
adding = np.frompyfunc(adding,2,1) ## (func,inputs_no,outputs_no)

print(adding([1,2,3],[4,5,6]))
```

```python
#Check if a Function is a ufunc
#If it is not a ufunc, it will return another type, like this built-in NumPy function for joining two or more arrays:
print(type(np.add))
print(type(np.concatenate))
print(type(np.where))
```

# Rounding Decimals
truncation 
fix 
rounding 
floor 
ceil 

```python
import numpy as np
arr = np.array([-3.1666, 3.6667])
#truncing
print(np.trunc([-3.1666, 3.6667]))

#fix
print(np.fix(arr))

#rounding
print(np.round(arr)) # >.5 ceil, .5< floor

#arround
print(np.around(arr,2)) #round to 2 decimal points

# ceil & floor
print(np.ceil(arr),np.floor(arr))

```

# Logs


```python
import numpy as np
arr = np.arange(1, 10) # 1---9

print(np.log2(arr)) # base 2
print(np.log10(arr)) # base 10
print(np.log(arr)) # base e/ natural log

# learn log at any base by creating a new function
```

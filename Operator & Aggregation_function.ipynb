```python
import numpy as np
list1 = [1,2,3,4,5]
list2 = [6,7,8,9,10]
list3 = [[23,32,23,34],
         [45,45,36,31]]
list4 = [[1.1,2],
         [5,6.3],
         [7,9]]
list5 = [
    [[1,2],[5,6],[7,9]],
    [[4,4],[1,4],[5,6]]
    ]
arr1 = np.array(list1)
arr2 = np.array(list2)
arr1_2d = np.array(list3)
arr2_2d = np.array(list4)
arr_3d = np.array(list5)
```

```python
# operators are same as any language
# but you can directly manupulte values without a loop
print(arr1)
print(arr1 + 5)
print(arr1 ** 2) #exponential
```

## <center>Some Aggregation functions</center>

```python
print(np.sum(arr1))
print(np.mean(arr1))
print(np.min(arr1))
print(np.max(arr1))
print(np.std(arr1)) #standard deviation
print(np.var(arr1)) #variance
```
